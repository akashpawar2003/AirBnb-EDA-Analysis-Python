Data Analysis

```python
#price distibution
sns.histplot(data=df,x="price")
```

```python
#identifying outliers in Price Columns

sns.boxplot(data=df,x="price")
```

```python
df=df[df['price']<1500]
sns.boxplot(data=df,x='price')
```

```python
sns.histplot(data=df,x='price',bins=100)
plt.title("Price Distributions")
plt.ylabel("Frequency")
plt.xlabel("Price")
plt.figure(figsize=(2,8))
plt.show()
```

```python
#availability_365 distibutions
sns.histplot(data=df,x='availability_365')
plt.title("availability_365 Distributions")
plt.ylabel("Frequency")
plt.xlabel("availability_365")
plt.figure(figsize=(2,5))
plt.show()
```

```python
# average price per neighbourhood_group
df.groupby("neighbourhood_group")["price"].mean().sort_values(ascending=False)
```

```python
#create a new column for price per bed
df['price per bed']=df['price']/df['beds']
```

```python
df['price per bed'].head(5)
```

```python
#Average Price per bed
df.groupby("neighbourhood_group")["price per bed"].mean().sort_values(ascending=False)
```
